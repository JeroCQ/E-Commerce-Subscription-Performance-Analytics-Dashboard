```
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random

# Set random seed for reproducibility
np.random.seed(42)

print("Generating Data for Power BI Project...")

# 1. Dim_Product (Specialty Coffee & Equipment)
products = {
    'Product_ID': ['P001', 'P002', 'P003', 'P004', 'P005', 'P006'],
    'Product_Name': ['Manual Drum Roaster', 'Green Beans - Geisha 1kg', 'Green Beans - Bourbon 1kg', 'Espresso Machine', 'Precision Scale', 'Burr Grinder'],
    'Category': ['Machinery', 'Raw Materials', 'Raw Materials', 'Machinery', 'Accessories', 'Accessories'],
    'Cost_Price': [250, 40, 25, 450, 15, 80],
    'List_Price': [499, 90, 55, 899, 45, 180]
}
dim_product = pd.DataFrame(products)

# 2. Dim_Customer
num_customers = 500
customer_ids = [f'C{str(i).zfill(4)}' for i in range(1, num_customers + 1)]
countries = ['Colombia', 'USA', 'Mexico', 'Spain', 'Thailand', 'UK']
signup_dates = pd.date_range(start='2024-01-01', end='2026-05-31', freq='D').to_list()

dim_customer = pd.DataFrame({
    'Customer_ID': customer_ids,
    'Country': np.random.choice(countries, num_customers, p=[0.3, 0.4, 0.1, 0.1, 0.05, 0.05]),
    'Signup_Date': [random.choice(signup_dates) for _ in range(num_customers)]
})

# 3. Fact_Sales_Orders
num_orders = 2500
order_dates = pd.date_range(start='2024-02-01', end='2026-06-20', freq='H').to_list()

fact_sales = pd.DataFrame({
    'Order_ID': [f'ORD{str(i).zfill(5)}' for i in range(1, num_orders + 1)],
    'Order_Date': [random.choice(order_dates).strftime('%Y-%m-%d') for _ in range(num_orders)],
    'Product_ID': np.random.choice(dim_product['Product_ID'], num_orders, p=[0.1, 0.3, 0.3, 0.05, 0.15, 0.1]),
    'Customer_ID': np.random.choice(customer_ids, num_orders),
    'Quantity': np.random.randint(1, 5, num_orders),
    'Channel': np.random.choice(['Web', 'Retail Partner', 'App'], num_orders, p=[0.6, 0.2, 0.2])
})

# 4. Fact_Subscriptions (AI Roasting Profile SaaS)
# Not all customers subscribe. Let's say 200 do.
sub_customers = random.sample(customer_ids, 200)
sub_starts = []
sub_ends = []
statuses = []

for cust in sub_customers:
    start = random.choice(pd.date_range(start='2024-03-01', end='2026-05-01', freq='D'))
    sub_starts.append(start.strftime('%Y-%m-%d'))

    # Simulate Churn: 30% chance they churned
    if random.random() < 0.30:
        duration_days = random.randint(30, 365)
        end = start + timedelta(days=duration_days)
        sub_ends.append(end.strftime('%Y-%m-%d'))
        statuses.append('Churned')
    else:
        sub_ends.append(pd.NA) # Active subscriptions have no end date yet
        statuses.append('Active')

fact_subs = pd.DataFrame({
    'Subscription_ID': [f'SUB{str(i).zfill(4)}' for i in range(1, len(sub_customers) + 1)],
    'Customer_ID': sub_customers,
    'Start_Date': sub_starts,
    'End_Date': sub_ends,
    'Monthly_Price': [29.99] * len(sub_customers),
    'Status': statuses
})

# Export to CSV
dim_product.to_csv('Dim_Product.csv', index=False)
dim_customer.to_csv('Dim_Customer.csv', index=False)
fact_sales.to_csv('Fact_Sales_Orders.csv', index=False)
fact_subs.to_csv('Fact_Subscriptions.csv', index=False)

print("Success! 4 CSV files have been generated.")
```
